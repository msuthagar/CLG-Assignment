```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data_path = 'bookks.csv'
df = pd.read_csv(data_path)

# Display the first few rows of the dataframe
print("Initial DataFrame:")
print(df.head())

# Handle missing values
df = df.dropna()

# Remove duplicates
df = df.drop_duplicates()

# Standardize text data
df['title'] = df['title'].str.title()
df['author'] = df['author'].str.title()

# Display the cleaned dataframe
print("\nCleaned DataFrame:")
print(df.head())

# Distribution of ratings
plt.figure(figsize=(10, 6))
sns.histplot(df['rating'], bins=20, kde=True)
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()

# Most popular genres
plt.figure(figsize=(10, 6))
popular_genres = df['genre'].value_counts().head(10)
sns.barplot(x=popular_genres, y=popular_genres.index, palette='viridis')
plt.title('Top 10 Genres')
plt.xlabel('Number of Books')
plt.ylabel('Genre')
plt.show()

# Most prolific authors
plt.figure(figsize=(10, 6))
prolific_authors = df['author'].value_counts().head(10)
sns.barplot(x=prolific_authors, y=prolific_authors.index, palette='magma')
plt.title('Top 10 Authors by Number of Books')
plt.xlabel('Number of Books')
plt.ylabel('Author')
plt.show()
```
